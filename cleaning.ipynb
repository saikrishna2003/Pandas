```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Customer Call List.xlsx")
```

```python
df.head()
```

```python
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns="Not_Useful_Column", errors='ignore')
df
```

```python
df.loc[:,'Last_Name'] = df['Last_Name'].str.strip(".../|_")
```

```python
df
```

```python
df.loc[:,'Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
df.loc[:,'Phone_Number']=df['Phone_Number'].replace('Na', np.nan)
```

```python
df.loc[:,'Phone_Number'] = df['Phone_Number'].fillna('')
```
